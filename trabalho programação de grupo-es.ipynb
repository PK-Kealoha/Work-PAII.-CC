# Qual é a duração média das músicas?

```python
mean_df = df['duration_ms'].mean()
print(mean_df)
```

# Converte os milissegundos em minutos

```python
#1s = 1000 milissegundos
#60s (1)- 60000 milissegundos
    #(2)  120000 milissegundos
    #(3)  180000 milissegundos
x_min = 1 * mean_df/60000
x_min
```

```python
df["duration_min"] = df["duration_ms"]/60000
df
```

# Quantas músicas tem a duração maior que 240000 milissegundos (4 minutos)?

```python
booleans = []
for w in df.duration_ms:
    if w >= 240000:
        booleans.append(True)
    else:
        booleans.append(False)
df.iloc[booleans]
```

# Quantas músicas tem a duração menor que 180000 milissegundos (3 minutos)?

```python
booleans = []
for w in df.duration_ms:
    if w <= 180000:
        booleans.append(True)
    else:
        booleans.append(False)
df.iloc[booleans]
```
